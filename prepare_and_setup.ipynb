## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/ILO-Youth-Unemployment/generated_data/'
input_file = 'transformed_data.csv'
```

```python
df = pd.read_csv(input_path + input_file, sep=',', encoding='utf-8')
df
```

The above data was transformed by *ILO-Youth-Unemployment* repository[1] which includes youth unemployment rate (% of total labor force ages 15-24) for OECD countries between 1991 to 2019.

The data originally collected International Labour Organization, ILOSTAT database and provided by World Bank [2]. 

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python

selected_cols = ['1991', '1992', '1993','1994','1995','1996',
           '1997','1998','1999','2000','2001','2002','2003','2004','2005',
           '2006','2007','2008','2009','2010','2011','2012','2013','2014',
           '2015','2016','2017','2018','2019']
# set figure size
plt.figure(figsize=(12,7))
x = list(range(1991,2020))
y = list(df[selected_cols].iloc[0])
plt.plot(x, y)

plt.xlabel('Year', fontsize=12)
# Set the y axis label of the current axis.
plt.ylabel('Unemployment rate (%)', fontsize=12)
# Set a title of the current axes.
plt.title('Youth unemployment rate in Australia in 2019', fontsize=17)
plt.ylim((0, 25))   # set the ylim to bottom, top

plt.savefig('img/Youth unemployment rate in Australia.pdf', bbox_inches='tight')
plt.show()
```

The above figure shows percentage of unemployed young people (aged 15-24) in Australia decreased from a high of 17.5 percent in 1991 to 8.8 percent in 2008. Unemployment increased to 13.3 percent in 2014 but decreased slightly up until 2019 (12 percent).

## Insight

**Around 12 percent of Australians (aged 15- 24) were unemployed and seeking employment in 2019.**

```python
value = float(df['2019'].iloc[0])
print('Around {} percent of Australians (aged 15- 24) were unemployed and seeking employment in 2019.'.format(round(value, 0)))
```

## Conclusion

We extract insight "Around 12.0 percent of Australians (aged 15- 24) were unemployed and seeking employment in 2019." from the transformed data. The transformed data generated by [1] that it contains information about Youth unemployment rate across OECD countries. According to the data, the youth unemployment rate in Australia has been relatively decreased between 1991 to 2019. The original data source of the transformed data is available at World Bank [2].
